```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
data=pd.read_csv("../datasets/GroceryStoreDataSet.csv", header=None)
data
```

```python
transactions=[]
items=data[0].values

for i in range(0, len(items)):
    transactions.append(items[i].split(","))

te=TransactionEncoder()
te_data= te.fit_transform(transactions)

dataTE=pd.DataFrame(data=te_data, columns=te.columns_, dtype=int)
dataTE
```

```python
support = apriori(dataTE, min_support = 0.15, use_colnames = True) #verbose = 1
support.sort_values(by="support",  ascending=False, inplace=True)
support
```

```python
df_ar = association_rules(support, metric = "confidence", min_threshold = 0.6)
df_ar
```

```python
df_ar2 = association_rules(support, metric="lift", min_threshold=2)
df_ar2
```
